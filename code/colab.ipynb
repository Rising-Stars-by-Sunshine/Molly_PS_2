```python
import numpy as np
import nashpy as nash

boy = np.array([[2, 0],   
               [0, 3]]) #row player

girl = np.array([[3, 0],
              [0, 2]])  #column player

game = nash.Game(boy,girl)

equilibria = game.support_enumeration()
for eq in equilibria:
  print(eq)
```

3. Nash equilibrium solution 

The pure strategy Nash equilibria occur when both players select the same option. In this case, there are two pure Nash equilibria: (Opera, Opera) and (Football, Football). If both players choose opera, neither player has an incentive to switch to football because the payoff for opera is higher for both players. The same reasoning applies to the (Football, Football) equilibrium.

3. Extreme equilibrium


The extreme equilibrium occurs when both players choose to defect, resulting in a payoff of (0, 0). This outcome is considered extreme because it is the worst possible collective outcome compared to the joint outcome of mutual cooperation (10, 10). In this equilibrium, neither player has an incentive to unilaterally change their strategy, as switching to cooperation would only benefit the other player.
